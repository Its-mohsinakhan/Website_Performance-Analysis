```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv("data-export.csv")
```

```python
df.head()
```

```python
df.columns=df.iloc[0]
```

```python
df=df.drop(index=0).reset_index(drop=True)
```

```python
df.columns=["channel group","Datehour","Users","Sessions","Engaged sessions","Average engagementtime persession","Engaged sessions per user","Events per session","Engagement rate","Event count"]
```

```python
df
```

```python
df.info()
```

```python
df["Datehour"]=pd.to_datetime(df["Datehour"],format="%Y%m%d%H",errors="coerce")
```

```python
numeric_cols=df.columns.drop(["channel group","Datehour"])
df[numeric_cols]=df[numeric_cols].apply(pd.to_numeric,errors="coerce")
df["Hour"]=df["Datehour"].dt.hour
```

```python
df.info()
```

```python
df.describe()
```

## sessions and user overtime

```python
sns.set(style="whitegrid")
```

```python
plt.figure(figsize=(80,80))
df.groupby("Datehour")[["Users","Sessions"]].sum().plot()
plt.title("sesssions and overtime")
plt.xlabel("Datehour")
plt.ylabel("count")
plt.show()
```

## Total users by Channel

```python
plt.figure(figsize=(10,5))
sns.barplot(data=df,x="channel group",y="Users")
plt.title("Total users by channel")
plt.xlabel("Channel group")
plt.ylabel("users")
plt.show()
```

## Average engagement time per channel

```python
plt.figure(figsize=(10,5))
sns.barplot(data=df,x="channel group",y="Average engagementtime persession",estimator=np.mean,palette="magma")
plt.title("Average engagement time")
plt.xticks(rotation=45)
plt.show()
```

## engagement rate per channel

```python
plt.figure(figsize=(10,5))
sns.boxplot(data=df,x="channel group",y="Engagement rate",palette="magma")
plt.title("Engagement rate per Channel")
plt.xticks(rotation=45)
plt.show()
```

## engagement and non engagement sessions

```python
session_df=df.groupby("channel group")[["Sessions","Engaged sessions"]].sum().reset_index()
session_df["non-engaged"]=session_df["Sessions"]-session_df["Engaged sessions"]
session_df_melted=session_df.melt(id_vars="channel group",value_vars=["Engaged sessions","non-engaged"])
```

```python
plt.figure(figsize=(10,5))
sns.barplot(data=session_df_melted,x="channel group",y="value",hue="variable")
plt.title("Engaged and non Engaged session")
plt.xticks(rotation=45)
plt.show()
```

## traffic by hour per channel

```python
heatmap_data=df.groupby(["Hour","channel group"])["Sessions"].sum().unstack().fillna(0)

plt.figure(figsize=(12,6))
sns.heatmap(heatmap_data,cmap="YlGnBu",linewidths=.5,annot=True,fmt=".0f")
plt.title("Traffic by hour and channel")
plt.xlabel("channel group")
plt.ylabel("hoiur of day")
plt.show()
```

## engagement rate ov er session overtime

```python
df_plot=df.groupby("Datehour")[["Engagement rate","Sessions"]].mean().reset_index()
plt.figure(figsize=(10,5))
plt.plot(df_plot["Datehour"],df_plot["Engagement rate"],label="Engagement rate",color="green")
plt.plot(df_plot["Datehour"],df_plot["Sessions"],label="Sessions",color="blue")
plt.title("Engagement Rate vs Session overtime")
plt.legend()
plt.xlabel("Datehour")
plt.show()
```
